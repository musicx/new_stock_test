```python
stocks = []
with open('c:\\Users\\mUSicX\\Documents\\TdxCloud\\export\\20210315.txt', 'r') as f:
    while True:
        l = f.readline().strip()
        if not l:
            break
        stocks.append(l)
    
stocks[:10]
```

```python
ff = open('c:\\Users\\mUSicX\\Documents\\TdxCloud\\export\\crawl20210315.txt', 'w')
for s in stocks[1:]:
    try:
        r=requests.get('http://basic.10jqka.com.cn/{}'.format(s), headers=ua)
        r.encoding='gbk'
        soup = BeautifulSoup(r.text, 'lxml')
        item = []
        item.append(s)
        core = soup.find('span', class_='core-view-text')
        if core is not None:
            item.append(core.text.strip())
        else:
            item.append('无亮点')
        main = soup.find('span', class_='main-bussiness-text')
        if main is not None:
            item.append(main.text.strip())
        else:
            item.append('无主营')
        concept= soup.find('div', class_='newconcept')
        cpt_text = []
        for cpt in concept.find_all('a', class_='newtaid') :
            if cpt.em is not None:
                cpt_text.append('{}*'.format(cpt.text))
            elif '详情' not in cpt.text:
                cpt_text.append(cpt.text)
        if len(cpt_text) > 0:
            item.append(','.join(cpt_text))
        else:
            item.append('无概念')
        line = '\t'.join(item)
    except:
        print('{} error'.format(s))
        continue
    ff.write(line)
    ff.write('\n')
    ff.flush()
    print(line)
    time.sleep(1.5)
    
ff.close()
```
