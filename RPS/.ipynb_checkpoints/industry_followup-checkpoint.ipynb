```python
import numpy as np
import pandas as pd
import pymysql as mysql
```

```python
connection = mysql.connect(host='localhost',
                           user='musicx',
                           password='Eric2mysql',
                           db='mx_stock',
                           charset='utf8mb4',
                           cursorclass=mysql.cursors.DictCursor)


query_sql = 'select * from high_mmt where datediff(current_date(), `date`) < 8 and `level`=2'
industries = pd.read_sql_query(query_sql, connection)
industries
```

```python
industries['rank'] = industries.groupby('date')['score'].rank(ascending=False, na_option='bottom')
industries
```

```python
ind_pivot = industries.pivot(index='industry', columns='date', values=['score','rank'])
ind_pivot = ind_pivot.reindex(sorted(ind_pivot.columns, reverse=True), axis=1)
ind_pivot.fillna(0, inplace=True)
ind_pivot.sort_values(by=ind_pivot.columns[0], ascending=False, inplace=True)
```

```python
pd.options.display.max_rows = 1000
top_ind = ind_pivot.loc[(ind_pivot.score > 1).any(axis=1),:]
top_ind
```

```python
market = pd.read_sql_query('select * from market where market <> "科创" and datediff(current_date(), ipo_date) > 120', connection)
chosen=market.loc[market.industry.isin(top_ind.loc[top_ind.iloc[:,0]>1, :].index.tolist()), :]
chosen.industry.value_counts()
```

```python
chosen.code.to_csv('top_rps_blocks.txt', index=False)
```
