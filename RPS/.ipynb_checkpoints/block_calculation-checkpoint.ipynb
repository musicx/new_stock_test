```python
import pandas as pd
import QUANTAXIS as qtx
from datetime import datetime, timedelta
```

```python
raw = pd.read_csv("c:\\Users\\mUSicX\\Documents\\股票相关\\20200922SW.csv")
raw['c'] = raw.code.apply(lambda x: x[:6])
raw['mkt'] = raw.code.apply(lambda x: 0 if x.endswith('SZ') else 1)
raw['ind_cat1'] = raw.ind.apply(lambda x: x.split('--')[0])
raw['ind_cat2'] = raw.ind.apply(lambda x: '-'.join(x.split('--')[:2]))
#raw['ind_cat3'] = raw.ind.apply(lambda x: x.split('--')[2].replace('Ⅲ', ''))

#raw['ind_cat'] = raw.apply(lambda x: x.ind_cat1+'-'+x.ind_cat2, axis=1)
raw[:10]
```

```python
code_list = qtx.QA_fetch_stock_list()
finfo =  qtx.QA_fetch_stock_info(list(code_list.code))
finfo.info()
```

```python
def mtm_ind(data, col, N=12, M=6):
    mtm = (data[col] - qtx.REF(data[col], N)) / data[col]
    mtmma = qtx.MA(mtm, M)
    return pd.DataFrame({'MTM': mtm, 'MTMMA': mtmma})
```

```python
fcnd = qtx.QA_fetch_stock_day_adv(list(code_list.code),  #['000001', '601555', '601058'],  
                                  start=(datetime.today()-timedelta(days=3600)).strftime('%Y-%m-%d'), 
                                  end=datetime.today().strftime('%Y-%m-%d')).to_qfq()
fcnd = fcnd.month
fcnd.head(10)
```

```python
fmnt = fcnd.reset_index()
fmnt['month'] = fmnt.date.apply(lambda x : x.strftime('%Y-%m'))
fmnt.sort_values(by=['code', 'month'], inplace=True)
fmnt.head(20)
```

```python
guben = finfo['zongguben'].reset_index()
fgb = pd.merge(fmnt, guben, on='code')
fgb['open_sz'] = fgb['open'] * fgb['zongguben']
fgb['close_sz'] = fgb['close'] * fgb['zongguben']
fgb['change'] = fgb['close'] / fgb['close'].shift(1) - 1
fgb = fgb.loc[:, ['month', 'code', 'open_sz', 'close_sz', 'change']]
fgb.head(20)
```

```python
fgb = pd.merge(fgb, raw.loc[:, ['c', 'ind_cat2']], left_on='code', right_on='c')
fgb.head(20)
```

```python
block_size = fgb.groupby(['ind_cat2', 'month'])['open_sz', 'close_sz'].sum().reset_index()
block_size.head(10)
```

```python
block_size.rename(columns={'open_sz':'open_all', 'close_sz':'close_all'}, inplace=True)
block_size.head(10)
```

```python
fbk = pd.merge(fgb, block_size, on=['month', 'ind_cat2'])
fbk['wchg'] = fbk.close_sz / fbk.close_all * fbk.change
bk_chg = fbk.groupby(['ind_cat2', 'month'])['wchg'].sum()
bk_chg.head(10)
```

```python
find = qtx.QA_fetch_index_day_adv(['000001', '000906'], 
                                  start=(datetime.today()-timedelta(days=3600)).strftime('%Y-%m-%d'), 
                                  end=datetime.today().strftime('%Y-%m-%d'))
find = find.month
find.head(10)
```

```python
find.reset_index(inplace=True)
find['month'] = find.date.apply(lambda x : x.strftime('%Y-%m'))
find.sort_values(by=['code', 'month'], inplace=True)
find['chg'] = find.close / find['close'].shift(1) - 1
find.head(10)
```

```python
mchg = pd.merge(bk_chg.reset_index(), find.loc[find.code=='000001', ['month', 'chg']], on='month')
mchg = mchg.loc[mchg.month != '2010-12', :]
mchg.sort_values(['ind_cat2', 'month'], inplace=True)
mchg.head(10)
```

```python
mcov = mchg.groupby('ind_cat2')['wchg', 'chg'].cov().reset_index()
mbeta = mcov.loc[mcov['level_1'] == 'chg', ['ind_cat2', 'wchg']]
mvar = mchg.groupby('ind_cat2')['chg'].var().reset_index()
mbeta = pd.merge(mbeta, mvar, on='ind_cat2')
mbeta.head()
```

```python
mbeta['beta'] = mbeta.wchg / mbeta.chg
mbeta.sort_values('beta', ascending=False, inplace=True)
mbeta.to_excel('level2_beta.xls')
mbeta
```
