```python
candles = qtx.QA_fetch_stock_day_adv(['000001', '000005'], 
                                     start=(datetime.today()-timedelta(days=100)).strftime('%Y-%m-%d'), 
                                     end=(datetime.today()-timedelta(days=10)).strftime('%Y-%m-%d')).to_qfq()
candles.data
```

```python
def mtm_ind(data, col, N=12, M=6):
    mtm = (data[col] - qtx.REF(data[col], N)) / data[col]
    mtmma = qtx.MA(mtm, M)
    return pd.DataFrame({'MTM': mtm, 'MTMMA': mtmma})
```

```python
mtm=candles.add_func(mtm_ind, col='close', N=20).loc[:, ['MTM','MTMMA']]
mtm
```

```python
mtm['rank'] = mtm.groupby('date')['MTM'].rank()
```

```python
mtm.loc[mtm['rank']<350,:]
```

```python
raw = pd.read_csv("c:\\Users\\mUSicX\\Documents\\股票相关\\20200922SW.csv")
raw['c'] = raw.code.apply(lambda x: x[:6])
raw['mkt'] = raw.code.apply(lambda x: 0 if x.endswith('SZ') else 1)
raw['ind_cat1'] = raw.ind.apply(lambda x: x.split('--')[0])
raw['ind_cat2'] = raw.ind.apply(lambda x: x.split('--')[1].replace('Ⅱ', ''))
raw['ind_cat3'] = raw.ind.apply(lambda x: x.split('--')[2].replace('Ⅲ', ''))

raw['ind_cat'] = raw.apply(lambda x: x.ind_cat1+'-'+x.ind_cat2, axis=1)
raw[:10]
```

```python
industry = raw.loc[:, ['c', 'ind_cat']].set_index('c').to_dict()
industry
```

```python
mtm = mtm.reset_index()
mtm['cat'] = mtm.apply(lambda x: industry['ind_cat'][str(x.code)], axis=1)
mtm.set_index(['date', 'code'])
mtm.head()
```

```python
group = mtm.loc[mtm['rank']<350,:].groupby(['date', 'cat'])
numbers = group['code'].count()

bases = mtm.groupby(['date', 'cat'])['code'].count()
```

```python
rps=pd.concat([pd.DataFrame(numbers).rename(columns={'code':'top'}), pd.DataFrame(bases).rename(columns={'code':'base'})], join ='inner', axis=1)
rps['score'] = rps.top / rps.base * rps.top
rps = rps.reset_index()
rps['rank'] = rps.groupby('date')['score'].rank(ascending=False, na_option='bottom')
rps.head(20)
```

```python
rps = rps.sort_values(['date', 'rank'])
```

```python
ind_pivot = rps.pivot(index='cat', columns='date', values=['score','rank'])
ind_pivot = ind_pivot.reindex(sorted(ind_pivot.columns, reverse=True), axis=1)
ind_pivot.fillna(0, inplace=True)
ind_pivot.sort_values(by=ind_pivot.columns[0], ascending=False, inplace=True)
top_ind = ind_pivot.loc[(ind_pivot.score > 1).any(axis=1),:]
top_ind
```

```python
top_ind.to_excel('rps_top_cat.xls')
```
