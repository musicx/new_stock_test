```python
import QUANTAXIS as qa
import numpy as np
import pandas as pd
```

```python
import re
```

```python
stocks = qa.QA_fetch_stock_list_adv().code.tolist()
stocks
```

```python
candles = qa.QA_fetch_stock_day_adv('300482', start = '2018-01-01', end='2018-07-13')
candles.data.tail(10)
```

```python
da = candles.data
da['rh'] = da.high.rolling(50).max()
```

```python
da['rl'] = da.low.rolling(50).min()
```

```python
da.tail(20)
```

```python
def double_digits(nin):
    ps = str(nin).replace('.', '')
    r1 = re.search(r'(\d)\1$', ps)
    r2 = re.search(r'(\d{2})\1', ps)
    # r3 = re.search(r'(\d)\1$', ps)
    return True if r1 or r2 else False
    
double_digits(55.77)
```

```python
mh = da.high.apply(double_digits) & (da.high == da.rh)
da.loc[mh, :]
```

```python
lh = da.low.apply(double_digits) & (da.low == da.rl)
da.loc[mh | lh, :]
```

```python
found = []
for stock in stocks:
    candles = qa.QA_fetch_stock_day_adv(stock, start = '2017-10-01', end='2018-07-13')
    if not candles:
        continue
    da = candles.data
    da['rh'] = da.high.rolling(50).max()
    da['rl'] = da.low.rolling(50).min()
    mh = da.high.apply(double_digits) & (da.high == da.rh)
    lh = da.low.apply(double_digits) & (da.low == da.rl)
    found.append(da.loc[mh | lh, :])
    
full = pd.concat(found, axis=0)
```

```python
full.loc[full.date>'2018-06-15', :]
```

```python
full.head()
```

```python
fs = full.reset_index(drop=True)
fs['hh'] = fs.high == fs.rh
fs['hl'] = fs.low == fs.rl
fh = fs.loc[fs.date>'2018-07-05',['date','code','hh', 'hl','high', 'low']]
fh
```

```python
fh = fs.loc[fs.date>'2018-07-05',['date','code','hh', 'hl','high', 'low']]
fh.loc[fh.hl, :].sort_values('date')
```
