```python
def prepare_days(stock):
    ca = qa.QA_fetch_stock_day_adv(stock,  start = '2018-01-01', end='2018-07-14').to_qfq().data
    ca['s1'] = ca.open / ca.preclose - 1
    ca['s2'] = ca.close / ca.open - 1
    days = ca.loc[:, ['s1', 's2']].stack()
    days.name = stock
    return days.reset_index(level=1, drop=True)
```

```python
full = pd.concat([prepare_days(x) for x in stocks[:10]], axis=1).reset_index(level=1, drop=True).fillna(0)
full.tail(30)
```

```python
cor =full.corr()
cor
```

```python
cc = cor.stack().reset_index()
cc['rr'] = cc.groupby('level_0')[0].transform(pd.DataFrame.rank, ascending=False)
cc
```

```python
cc['rr'] = pd.rank
```

```python
cdict = {}
```

```python
def insert(x):
    y = x.values
    if y[0]<y[1]:
        cdict['{}_{}'.format(y[0], y[1])] = y[2]
    return 0
    
cc.apply(insert, axis=1)
```
