```python
from pymodline.labeled import forest_trainer as ft
mod, vals = ft.train_random_forest(dev, 'GOUP', ignores=['date', 'code'], num_trees=1000, samples_split=20, sample_leaf=5, magic_number=518)
print('model trained')

for name, imp in zip(vals, mod.feature_importances_):
    print('{}: {}'.format(name, imp))

ft.save_random_forest(mod, 'goup2', feature_names=vals)
print('model saved')
```

```python
import pymodline.eval.score_evaluator as eva
```

```python
pred = ft.score(val, mod, vals, 'GOUP', ['date', 'code', 'GOUP'])
pred.sort_values('prob_of_1', ascending=False, inplace=True)
pred.to_csv('D:/Work/Python/MyCode/stock_model/data/rf_predict_val2.csv')
print('score predicted')

data = eva.prepare_data(pred, 'GOUP')
sl = eva.check_single_score(data, 'prob_of_1', 'GOUP')
op = eva.check_operation_points(sl, points=1000)
ocm = eva.analyze_performance(op, 1)
# rpt.create_tabular_report('../data/rf_report.csv', ocm)
ocm.to_csv("D:/Work/Python/MyCode/stock_model/data/rf_report2.csv")
print('done')

fut_pred = ft.score(test, mod, vals, 'GOUP', ['date', 'code'])
fut_pred.sort_values('prob_of_1', ascending=False, inplace=True)
fut_pred.to_csv('D:/Work/Python/MyCode/stock_model/data/rf_predict2.csv')
```
