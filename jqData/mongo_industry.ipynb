```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client['jqdata']
coll = db['industry']
```

```python
cursor = coll.find({'code': '600001', 'date': {'$gt': '2021-03-01'}}).sort('date_stamp', -1)
```

```python
for c in cursor:
    print(c)
```

```python
data = pd.DataFrame(
            [
                item for item in coll
                .find({'code': {
                    '$in': ['600000', '600004']
                },
                'date': {
                    '$in': ['2021-03-03']
                }
            },
                {"_id": 0},
                batch_size=10000)
            ]
        )
data
```

```python
ind = data.pivot(index='code', columns='type', values='ind_name')
ind['sw'] = ind.apply(lambda x: '-'.join([x['sw1'], x['sw2'], x['sw3']]), axis=1)
ind
```

```python
data.set_index('code', inplace=True)
data
```
